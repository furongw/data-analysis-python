```python
all_data = [['ho','yyg','fr','kjw','xxy'],['cr','xeer','lk','jpf','ytc']]
```

以下两种相等

```python
names_of_interest = []
for names in all_data:
    enough_es = [name for name in names if name.count('e') >=2]
    names_of_interest.extend(enough_es)
```

# 函数

函数是对象

```python
states = ['  Alabama','Georgia!','Georgia','georgia','FlOrIda','south    carolina##','west balaa?']
```

```python
import re
def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub('[!#?]','',value)
        value = value.title()
        result.append(value)
    return result
```

```python
clean_strings(states)
```

```python
def remove_punctuation(value):
    return re.sub('[!#?]','',value)

clean_ops = [str.strip, remove_punctuation,str.title]

def clean_strings(strings, ops):
    result = []
    for value in strings:
        for function in ops:
            value = function(value)
        result.append(value)
    return result
```

```python
clean_strings(states,clean_ops)
```

```python
for x in map(remove_punctuation, states):
    print(x)
```

## 生成器

生成器是构造新的可遍历对象的一种非常简洁的方式。普通函数执行并一次返回单个结果，而生成器则惰性地返回一个多结果序列，在每一个元素产生之后暂停，直到下一个请求，如需创建一个生成器，只需要在函数中将返回关键字return替换为yield关键字：

```python
def squares(n=10):
    print('Generate squares from 1 to {0}'.format(n**2))
    for i in range(1,n+1):
        yield i**2
```

当实际调用生成器时，代码并不会立即执行

```python
gen = squares()
gen
```

直到请求生成器中的元素时，它才会执行它的代码：

```python
for x in gen:
    print(x, end = ' ')
```

生成器表达式

```python
gen = (x**2 for x in range(100))
gen
```

在很多情况下，生成器表达式可以作为函数参数用于替代列表推导式：

```python
sum(x**2 for x in range(100))
```

```python
dict((i,i**2) for i in range(3))
```

itertools模块

标准库中地itertools模块适用于大多数数据算法的生成器集合

```python
import itertools
first_letter  = lambda x:x[0]
names = ['xeer','furong','yyg']
for letter, names in itertools.groupby(names,first_letter):
    print(letter,list(names))
```

## 错误和异常处理

```python
def attempt_float(x):
    try:
        return float(x)
    except (TypeError, ValueError):
        return x
```

如果想处理一个异常，且无论是否报错都要执行则使用finally关键字

```python
f = open(path,'w')
try:
    f.write('trytry')
except:
    print('failed')
finally:
    f.close()
```

## 字节与Unicode文件

除了二进制模式，在打开文件时使用seek要当心。如果文件的句柄位置恰好在一个Unicode符号的字节中间时，后续的读取会导致错误

open时用encoding=可以将文件内容从Unicode编码转换为其他类型的编码

```python
with open (path, encoding='iso-8859-1') as f:
    print(f.read(10))
```
