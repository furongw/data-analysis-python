# 基本功能

## 重建索引

```python
obj = pd.Series([4.5,7.2,-5.3,3.6],index = ['d','b','a','c'])
```

```python
obj
```

```python
obj2 = obj.reindex(['a','b','c','d','e'])
```

```python
obj2
```

对于顺序数据，比如时间序列，在重建索引时可能会需要进行插值或填值。method可选参数允许我们使用诸如ffill等方法在重建索引时插值ffil方法会将数值前向填充

```python
obj3 = pd.Series(['blue','purple','yellow'],index = [0,2,4])
```

```python
obj3
```

```python
obj3.reindex(range(6),method = 'ffill')
```

reindex既可以改变行索引又可以改变列索引

```python
frame = pd.DataFrame(np.arange(9).reshape((3,3)),index = ['a','c','d'],columns=['Ohio','Texas','California'])
```

```python
frame
```

```python
states = ['Xeer','Texas','California']
frame.reindex(columns = states)
```

更简洁的方法

```python
frame.loc[['a','b','c','d'],states]
```

将对象相加时，如果存在某个索引对不相同，则返回结果的索引将是索引对的并集

两个索引化对象相加时可以使用特殊填充值

```python
df1 = pd.DataFrame(np.arange(12).reshape((3,4)),columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20).reshape((4,5)),columns=list('abcde'))
```

```python
df2.loc[1,'b'] = np.nan
```

```python
df1
```

```python
df2
```

```python
df1+df2
```

```python
df1.add(df2,fill_value = 0)
```

```python
1/df1
```

```python
df1.rdiv(1)
```

默认情况下，DataFrame和Series的数学操作中会将Series的索引和DataFrame的列进行匹配，并广播到各行

```python
series = frame.iloc[0]
```

```python
frame
```

```python
series
```

```python
frame-series
```

如果想改为在列上进行广播，在行上匹配，必须使用算术方法的一种

```python
series3 = frame['d']
```

```python
frame
```

## 函数应用和映射

Numpy的通用函数（逐元素数组方法）对pandas对象也有效

```python
frame = pd.DataFrame(np.random.randn(4,3),columns=list('bde'),index = ['Utah','Ohio','Texas','Oregon'])
```

```python
frame
```

```python
np.abs(frame)
```

将函数应用到一行或一列的一维数组上：DataFrame的apply方法

```python
f = lambda x:x.max()-x.min()
```

```python
frame.apply(f)
```

```python
frame.apply(f,axis = 'columns')
```

传递给apply的函数并不一定要返回一个标量值，也可以返回带有多个值的Series

```python
def f(x):
    return pd.Series([x.min(),x.max()],index = ['min','max'])
```

```python
frame.apply(f)
```

逐元素的python函数可以用applymap

```python
format = lambda x:'%.2f' % x
```

```python
frame.applymap(format)
```

使用applymap作为函数名是因为Series有map方法，可以将一个逐元素的函数应用到Series上

```python
frame['e'].map(format)
```

## 排序和排名

按索引排序

```python
frame.sort_index()
```

```python
frame.sort_index(axis = 1)
```

按照值排序

```python
obj = pd.Series([4,np.nan,7,np.nan,-3,2])
```

```python
obj.sort_values()
```

```python
frame = pd.DataFrame({'b':[4,7,-3,2],'a':[0,1,0,1]})
```

```python
frame
```

```python
frame.sort_values(by = 'b')
```

排名是指对数组从1到有效数据点总数分配名次的操作。默认情况下，rank通过将平均排名分配到每个组来打破平级关系。

```python
obj = pd.Series([7,-5,7,4,2,0,4])
```

```python
obj.rank()
```

按照观察顺序

```python
obj.rank(method = 'first')
```

```python
obj.rank(ascending = False, method = 'max')
```

DataFrame可以对行或列计算排名

```python
frame = pd.DataFrame({'b':[4.3,7,-3,2],'a':[0,1,0,-1],'c':[-2,5,8,-2.5]})
```

```python
frame
```

```python
frame.rank(axis = 'columns')
```

## 相关性与协方差


```python
import pandas_datareader.data as web
```

```python
all_data = {ticker:web.get_data_yahoo(ticker) for ticker in ['AAPL','IBM','MSFT','GOOG']}
```

```python
all_data
```

```python
price = pd.DataFrame({ticker:data['Adj Close'] for ticker,data in all_data.items()})
```

```python
price
```

```python
volume = pd.DataFrame({ticker:data['Volume'] for ticker,data in all_data.items()})
```

```python
volume
```

```python
returns = price.pct_change()  # 计算股价的百分比
```

```python
returns
```

计算协方差

```python
returns['MSFT'].corr(returns['IBM'])  #计算的是两个Series中的重叠的非NA的、按索引对其的值的相关性
```

```python
returns['MSFT'].cov(returns['IBM'])  # 协方差
```

```python
returns.corr()
```

```python
returns.cov()
```

```python
returns.corrwith(returns.IBM)
```

```python
returns.corrwith(volume)
```
